##### 筆記：flag ＋ break 解題技巧
- flag 使用時機：flag應用在「需要辨識狀態」上，可以初始建立一個false/true flag，然後透過檢驗改變狀態。
- 注意事項：小心局部變數和全域變數的問題
- break 的使用時機：
    - break 功能：停止迴圈並跳出
    - 本題的應用：因為題目說「只要有一個true就好」，並且希望只返回一個「true」，所以為了避免列印出所以字符的狀態，`break`的使用可以只要一個true就跳出回圈。
- `if .isalnum():` 為什麼不用加上 `== True`？
    - 因為 `.isalnum()`本身會回傳布林值，所以`== True`可以省略
    - 注意：如果條件是false，不會是`== False`而是`if not i.isalnum():`

- 示範運作：
```python
def test(s):
    test_flag = False
    for i in s:
        if i.isalnum():
            test_flag = True
            break
    return test_flag
```
    - 基本上只要i有false的，就不會進入到if的條件裡

##### 筆記：全域變數 ＆ 局部變數
- 定義：
    - 全域變數：所有程式空間共同使用的變數，所有的def都可以訪問、更改，
        - 具備一般不建議def裡直接更改，原因是互相依存的關係會放讓此變數複雜度增加
        - def 內的呼叫方法：`global`+ 全域變數名稱
            ```python
            global_count = 0
            def test(s):
                global global_count # call global variable 
            ```
    - 局部變數：局部空間（例如def裡面）的變數，變數只能由聲名他的函示或塊裡被訪問、修改
        - 一般建議以局部變數進行修改，避免副作用或相互依賴
        - 示範：
            ```python 
            def test(s):
                count = 0 # local variable
                count += 1
            ```
    - 內置預設（built-in）：python 預設的函數名稱如abs, char...，在py中，不需要像c語言一樣手動預設，例如：
    ```c
    #include <stdio.h> // 為了print
    int main(){
        printf('hello');
    }
    ```

    ```python
    print('hello')  # 不需要任何import
    ```

- 為什麼剛剛的程式出錯？
    - 測試時正確，因為可能只運行一次，所以正確
    - submit時有多個case需要檢驗，多次調用下可能就會出錯
    - 小筆記：py的讀取會以LEGB規則執行：local -> enclosing（當有巢狀def時的外層） -> global -> built-in


- 實際測試 
```python

x = 10 # global variable

def test1():
    print(x) # result:10

def test2():
    x = 20 # set new local variable
    print(x) # result:20

def test3():
    print(x) # 這段會出錯，因為會先認為x應該是local，但是def內x並未被賦值
    x = 30 # 後面才賦值

def test4():
    global x # 明確告訴python要用全域的x
    print(x) #result:10

```

##### 筆記：解題方法論 － 累加器或狀態管理
- 核心概念：初始化 -> 迭代更新 -> 返回結果
- 經典的用法：
    - 布林值累加（像是本次的flag)
    ```python
        found = False        # 🏁 初始：假設沒找到
        for item in items:
            if condition(item):
                found = True # ✅ 更新：找到了！
                break    
    ```
    - 數值累加
    ```python
        count = 0           # 🏁 初始：從零開始
        for item in items:
            if condition(item):
                count += 1  # ➕ 更新：累加
    ```    
    - 容器累加
    ```python
        result = []         # 🏁 初始：空容器
        for item in items:
            if condition(item):
                result.append(item)  # 📦 更新：收集元素
    ```
    - 位置累加（雙指針）
    ```python
        left, right = 0, len(arr)-1  # 🏁 初始：兩端位置
        while left < right:
            # 🔄 更新：根據條件移動指針
            if condition():
                left += 1
            else:
                right -= 1
    ```    
    - 進階：多狀態累加
    ```python
        def analyze_array(nums):
            # 🏁 初始化多個狀態
            max_val = float('-inf')
            min_val = float('+inf') 
            sum_val = 0
            positives = []
            
            # 🔄 一次遍歷，同時更新所有狀態
            for num in nums:
                max_val = max(max_val, num)     # 更新最大值
                min_val = min(min_val, num)     # 更新最小值  
                sum_val += num                  # 更新總和
                if num > 0:
                    positives.append(num)       # 收集正數
            
            # 📊 返回綜合結果
            return {
                'max': max_val,
                'min': min_val, 
                'average': sum_val / len(nums),
                'positives': positives
            }
    ```
