````python
# ä½¿ç”¨flagè®Šæ•¸ - éŒ¯èª¤ï¼ï¼è®Šæ•¸ä½œç”¨åŸŸå•é¡Œ
s = "qA2"
number_flag = alpha_flag = digit_flag = lower_flag = upper_flag = False

def is_number(s):
    number_flag = False
    for i in s:
        if i.isalnum():
            number_flag = True 
            break
    return number_flag

def is_alphabet(s):
    for i in s:
        if i.isalpha():
            alpha_flag = True 
            break
    return alpha_flag

def is_int(s):
    for i in s:
        if i.isdigit():
            digit_flag = True 
            break
    return digit_flag

def is_lower(s):
    for i in s:
        if i.islower():
            lower_flag = True 
            break
    return lower_flag

def is_upper(s):
    for i in s:
        if i.isupper():
            upper_flag = True 
            break
    return upper_flag


print(is_number(s))
print(is_alphabet(s))
print(is_int(s))
print(is_lower(s))
print(is_upper(s))
````
